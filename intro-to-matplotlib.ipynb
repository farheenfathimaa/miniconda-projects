## matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in the jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# 2. Setup plot
fig,ax=plt.subplots(figsize=(10,10))

# 3. plot data
ax.plot(x,y)

# 4. customise plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. save and show 
fig.savefig("images/sample-plot.png")
```

## 2 options for subplots

```python
# subplot option 1
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(figsize=(10,5),nrows=2,ncols=2)

#plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplot option 2
fig,ax=plt.subplots(figsize=(10,5),nrows=2,ncols=2)
#plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### which one should you use? (pyplot vs matplotlib OO method?)
* when plotting something quickly, okay to use the pyplot method
* when plotting somthing more advanced,use the OO method

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
#pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             figsize=(10,6));
```

```python
#OO method mixed with pyplot
fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
ax.set_xlim([45,100]);
```

```python
# OO method from scratch

# create plot
fig, ax=plt.subplots(figsize=(10,6))

# plot data
scatter=ax.scatter(x=over_50['age'],
             y=over_50['chol'],
             c=over_50['target']);

# customize plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

#Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
over_50.head()
```

```python
#subplot of chol,age,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True)
# add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
# add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])
#customize both
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")
ax1.set(title="Heart Disease and Max Heart rate",
       xlabel="Age",
       ylabel="Heart rate")

# legend 
ax0.legend(*scatter.legend_elements(),title="Target");
ax1.legend(*scatter.legend_elements(),title="Target");

# horizontal line
ax0.axhline(over_50['chol'].mean(),
            linestyle='--',
            color='black');
ax1.axhline(over_50['thalach'].mean(),
            linestyle='--',
            color='black');

# add a title to the fig
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
#set the style
plt.style.use('seaborn-v0_8-whitegrid')
# OO method from scratch

# create plot
fig, ax=plt.subplots(figsize=(10,6))

# plot data
scatter=ax.scatter(x=over_50['age'],
             y=over_50['chol'],
             c=over_50['target'],
            cmap="plasma"); #this changes the colour scheme

# customize plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

#Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
#subplot of chol,age,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True)
# add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                   cmap="winter")
# add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                   cmap="winter")
#customize both
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

ax0.set_xlim([50,80]) #change x axis limits 

ax1.set(title="Heart Disease and Max Heart rate",
       xlabel="Age",
       ylabel="Heart rate")

ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# legend 
ax0.legend(*scatter.legend_elements(),title="Target");
ax1.legend(*scatter.legend_elements(),title="Target");

# horizontal line
ax0.axhline(over_50['chol'].mean(),
            linestyle='--',
            color='black');
ax1.axhline(over_50['thalach'].mean(),
            linestyle='--',
            color='black');

# add a title to the fig
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

```
